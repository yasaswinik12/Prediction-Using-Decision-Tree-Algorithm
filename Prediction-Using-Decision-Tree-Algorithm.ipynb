# Importing modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
```

```python
df = pd.read_csv(r'C:\Users\Yasaswini\Desktop\Iris.csv')
df = df.drop(df.columns[[0]], axis=1)
df.head()
```

# Relation between columns of the dataset

```python
sns.pairplot(df, hue='Species')
```

# Correlation between columns by using heatmap

```python
plt.figure(figsize=(10,11))
sns.heatmap(df.corr(),annot=True)
plt.plot()
```

We can see Petal Length and Petal Width are almost correlatted.

# Using boxplot to see how the “Species” is distributed with other columns.

```python
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
sns.boxplot(x="Species",y="SepalLengthCm",data=df)
plt.subplot(2,2,2)
sns.boxplot(x="Species",y="SepalWidthCm",data=df)
plt.subplot(2,2,3)
sns.boxplot(x="Species",y="PetalLengthCm",data=df)
plt.subplot(2,2,4)
sns.boxplot(x="Species",y="PetalWidthCm",data=df)
```

# Splitting the dataset

```python
x = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
y = df['Species'].values

(train_x, test_x, train_y, test_y) = train_test_split(x, y, train_size=0.7, random_state=1)
```

# Classification and measuring the accuracy

Classification

```python
dtc = DecisionTreeClassifier()
dtc.fit(train_x, train_y)
```

```python
y_pred_tree=dtc.predict(test_x)
print(y_pred_tree)
```

Accuracy

```python
print("\nAccuracy: ",dtc.score(test_x, test_y))
```

# Plotting the Graph

```python
from sklearn import tree
fig,ax=plt.subplots(figsize=(15,8))
tree.plot_tree(dtc,feature_names=list(df),class_names='Species',filled=True)
```
